```python
# .append() 사용법

numbers = [1,2,3,4,5]
numbers.append(6)
print(numbers)
```

```python
# [] 여기안에 넣어서 숫자는 넣을 수는 있는데 아쉽다.

numbers = [1,2,3,4,5,6]
numbers.append([7,8])
print(numbers)
```

```python
# 언패킹을 써서 풀리지 않는다.

numbers.append(*[7, 8])
```

```python
# extend 메소드를 활용하여 7, 8을 한꺼번에 넣을 수 있다.

numbers = [1,2,3,4,5,6]
numbers.extend([7,8])
print(numbers)
```

```python
# insert메소드 활용법1 : 임의로 숫자를 넣고 싶을 때 쓴다.

numbers = [1,2,3,4,5,6]
numbers.insert(1, 10)
print(numbers)
```

```python
# 그냥 마지막에 넣고 싶을 땐 이렇게 작성할 수도 있다.

numbers = [1,2,3,4,5]
numbers.insert(len(numbers), 12)
numbers
```

```python
# insert 활용법2

numbers.insert(2, [10, 11])
numbers
```

```python
# insert

numbers = [1,2,3,4,5,6]
numbers.insert(2, 11)
numbers.insert(2, 10)
print(numbers)
```

```python
# 슬라이스연산

numbers = [1,2,3,4,5]
numbers[1:3]
```

```python
# 슬라이스연산(요소를 추가함)

numbers = [1,2,3,4,5]
numbers[2:2] = [10, 11]
numbers
```

```python
# 슬라이스연산(요소를 잘라내고 갈아 끼움)
numbers = [1,2,3,4,5]
numbers[2:3] = [10, 11]
numbers
```

```python
numbers = [1,2,3,4,5]
numbers[len(numbers):] = [10]
numbers
```

```python
numbers = [1,2,3,4,5]
numbers[len(numbers):] = 10
```

```python
numbers = [1,2,3,4,5]
numbers[len(numbers):] = (20,30)
numbers
```

리스트에 새로운 요소 추가

append
리스트의 마지막에 새로운 요소 추가
extend
주어진 시퀀스 자료형 값의 요소들을 리스트의 마지막에 새로운 요소로 추가한다.
insert
리스트에서 첫 번째 인자로 주어진 인덱스의 위치에 두 번째 인자를 추가한다.
