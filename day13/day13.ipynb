```python
# .append() 사용법

numbers = [1,2,3,4,5]
numbers.append(6)
print(numbers)
```

```python
# [] 여기안에 넣어서 숫자는 넣을 수는 있는데 아쉽다.

numbers = [1,2,3,4,5,6]
numbers.append([7,8])
print(numbers)
```

```python
# 언패킹을 써서 풀리지 않는다.

numbers.append(*[7, 8])
```

```python
# extend 메소드를 활용하여 7, 8을 한꺼번에 넣을 수 있다.

numbers = [1,2,3,4,5,6]
numbers.extend([7,8])
print(numbers)
```

```python
# insert메소드 활용법1 : 임의로 숫자를 넣고 싶을 때 쓴다.

numbers = [1,2,3,4,5,6]
numbers.insert(1, 10)
print(numbers)
```

```python
# 그냥 마지막에 넣고 싶을 땐 이렇게 작성할 수도 있다.

numbers = [1,2,3,4,5]
numbers.insert(len(numbers), 12)
numbers
```

```python
# insert 활용법2

numbers.insert(2, [10, 11])
numbers
```

```python
# insert

numbers = [1,2,3,4,5,6]
numbers.insert(2, 11)
numbers.insert(2, 10)
print(numbers)
```

# 컴파일 언어와 인터프리터 언어의 변수의 정의와 대입

- 컴파일 언어 : 변수가 어느 주소에 저장, int num = 10을 붙여도 int num = 20을 붙이면 num은 10, 20이 들어감

- 인터프리터 언어 : 10이라는 곳에 num이라는 이름표를 붙임

그래서 인터프리터 언어에 10이라는 num 이름표를 20이라는 숫자에 이름표에 붙여버리면 10은 이름표가 떼져서 num은 20만 된다.

```python
# 변수 개념 최종 정리

num = 10 # 일단 num에 10을 지정
val = num # val은 10이 됨
num = 20 # num을 20으로 변경, 더이상 num은 10이 아님
```
