```python
# 연습문제1
# 리스트 'numbers'의 요소들을 차례대로 출력

numbers = [4, 17, 32, 3, 8, 13, 41, 27, 16, 1]

# for 구문과 range 함수로 리스트 numbers의 요소에 차례대로 접근
for i in range(len(numbers)):
    print(f'numbers[{i}] = ', numbers[i])
```

```python
# enumerate 함수 : 인자로 주어진 시퀀스 자료형 값에서 요소의 인덱스와 값의 쌍을 튜플로 생서애서 차례대로 추출할 수 있게 하는 
# enumerate 객체를 생성

numbers = [4, 17, 32, 3, 8, 13, 41, 27, 16, 1]

enumerate(numbers)
```

```python
# for 구문과 enumerate 함수로 리스트 numbers의 요소에 차례대로 접근
# index, item = (0,4)
# index, item = (1,17)
# index, item = (2,32)
# ...

for index, item in enumerate(numbers):
    print(f'numbers[{index}] = {item}')
```

```python
# 시퀀스 자료형 값에서 인덱스와 값을 추출

name = ['Kim','Park','Cho']

for index, name in enumerate(name, start=1):
    print(f'{index}. {name}')
```

```python
# 연습문제2
# 리스트 scores에서 80보다 큰 요소들의 인덱스와 값을 차례대로 출력한다.

scores = [88, 92, 76, 64, 80, 96, 84, 72, 68, 100]

# for 구문과 range 함수로 리스트의 요소를 차례대로 확인
for i in range(len(scores)):
    # 리스트 scores에서 인덱스가 i인 요소가 80보다 크면 출력
    if scores[i] > 80: #내 답안 : if i > 80:
        print(f'scores[{i}] = ', scores[i])
```

```python
# 연습문제2 다른풀이

# for 구문과 enumerate 함수로 리스트의 요소를 차례대로 확인
for index, item in enumerate(scores):
    # 변수 item이 80보다 크면 출력
    if item > 80:
        print(f'scores[{index}] = {item}')
```

```python
# 리스트 numbers의 요소들 중에서 짝수 번째 요소들을 차례대로 출력한다.

numbers = [4, 17, 32, 3, 8, 13, 41, 27, 16, 1]

for i in range(len(numbers)):
    if numbers[i] % 2 == 0: # == if i % 2 = 0
        print(f'scores[{i}] = ', numbers[i])
```

```python
# for 구문과 enumerate 함수로 리스트 numbers의 요소를 차례대로 확인
for index, item in enumerate(numbers):
    # 요소의 인덱스를 나타내는 변수 index가 홀수이면 변수 item을 출력
    # if index % 2 == 1:
    if index % 2: # 만약 짝수를 출력하고 싶다면 if index % 2 == 0을 직접 입력해야한다.
        print(item)
```

```python
# for 구문과 range 함수로 리스트 numbers의 요소들 중에서 인덱스가 홀수인 요소들을 출력
# i = 1,3,5,7,9

for i in range(1, len(numbers), 2):
    print(numbers[i])
```

```python
# for 구문과 슬라이스 연산으로 리스트 numbers의 요소들 중에서 인덱스가 홀수인 요소들을 출력

for item in numbers[1::2]:
    print(numbers[item])
```
