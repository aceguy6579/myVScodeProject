```python
# 연습문제2 : 리스트 취미에서 요소가 '콜라'라면 '소프트 드링크'로 출력하고 그렇지 않으면 요소를 그대로 출력한다.
# 선생님 답안5

favorites = ['커피', '차', '쥬스', '콜라', '우유']

# 빈 리스트 생성
copy = []

# for 구문으로 리스트 copy에 리스트 favorites를 복사
for item in favorites:
    copy.append('소프트 드링크' if item == '콜라' else item)
    
# 출력
print(copy) # 그냥 출력
print(*copy) # 언패킹 연산자 사용
print(*copy, sep=' / ') # 언패킹 연산자 사용 + 대행 연산자
print(*copy, end=' / ') # 언패킹 연산자 사용 + 대행 연산자
```

```python
# 연습문제2 : 리스트 취미에서 요소가 '콜라'라면 '소프트 드링크'로 출력하고 그렇지 않으면 요소를 그대로 출력한다.
# 선생님 답안6

# 리스트 표현식으로 리스트 favorites를 복사해서 새로운 표현식을 생성
copy = ['소프트 드링크' if item == '콜라' else item for item in favorites]
print(copy) # 그냥 출력
```

```python
# 사전형 메뉴에서 카페에서 판매하는 음료의 가격 중 1500원 이하인 음료들만 출력하자.
# 내 답안1

menu = {'커피': 2000, '쥬스': 2200, '콜라': 1500, '우유': 1200}

for key, value in menu.items():
    if value <= 1500:
        print(f'{key.title()}: {value}', end= '/ ')
```

```python
for value in menu.values():
    if value <= 1500:
        print(f'{value.title()} : ', value) # key.title은 되는데 value.title은 안된다.
```

```python
# 선생님 답안1

# for 구문으로 사전형 menu의 요소를 차례대로 확인
for key in menu:
# for key in menu.keys():
    # 키가 변수 key인 요소가 1500 이하이면 출력
    if menu[key] <= 1500:
        print(f'{key}: {menu.get(key)}', end= '/ ')
```

```python
# for 구문과 items 메서드로 사전형 menu의 요소를 차례대로 확인
for key, value in menu.items():
    # 변수 value가 1500 이하이면 출력
    if value <= 1500:
        print(f'{key}: {value:,}') # 1000단위마다 ,찍기
```
