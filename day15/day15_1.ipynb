```python
scores = {'korean' : 86, '수학': 97, '영어':68}

for key, value in scores.items():
    print(f'{key.title()}: {value}', end= '/ ')
```

```python
# 이름이 '정확'이라는 학생 질문:
# 선생님 keys 나 values 메서드가
# 리스트랑 비슷한거라고 하셨자나요
# 리스트는 아니니까 append extend 메서드 같은건
# 적용이 안되나요?
scores = {'korean' : 86, '수학': 97, '영어':68}
values = scores.values()
values.append('') # append는 메서드로 새로운 요소를 추가할 수 없다. 리스트와 비슷하지만 적용이 안된다고 한다.
# keys, values는 view 객체이다. 그래서 리스트가 적용되아하는 append가 안먹는다
```

```python
# 사전형 값 생성
scores = {'korean' : 86, '수학': 97, '영어':68}
# 사전형의 values 메서드로 dict_values 객체 생성
values = scores.values()
# 사전형 값을 리스트로 변환
values = list(values)
# 리스트의 append 메서드로 새로운 요소 추가
values.append(72)
# 출력
print('values = ', values)
```

```python
# 연습문제1 : 리스트 '취미'에서 요소가 '커피', '차'이면 요소의 인덱스와 값을 출력한다.
# 내 답안

favorites = ['커피', '차', '쥬스', '콜라', '우유']

for i in range(len(favorites)):
    if favorites[i] == '커피' or favorites[i] == '차':
        print(f'{i}.', favorites[i])
```

```python
# 연습문제1 : 리스트 '취미'에서 요소가 '커피', '차'이면 요소의 인덱스와 값을 출력한다.
# 선생님 답안1

favorites = ['커피', '차', '쥬스', '콜라', '우유']

# for 구문과 range 함수로 리스트 favorites의 요소를 차례대로 확인
for i in range(len(favorites)):
    # 인덱스가 변수 i인 요소가 "커피"이거나 "차"이면 출력
    if favorites[i] == '커피' or favorites[i] == '차':
        print(f'{i}.', favorites[i])
```

```python
# 연습문제1 : 리스트 '취미'에서 요소가 '커피', '차'이면 요소의 인덱스와 값을 출력한다.
# 선생님 답안2

# for 구문과 enumerate 함수로 리스트 favorites의 요소를 차례대로 확인
for index, item in enumerate(favorites):
    # 인덱스가 변수 i인 요소가 "커피"이거나 "차"이면 출력
    if favorites[index] == '커피' or favorites[index] == '차':
        print(f'{index}. {item}')
```

```python
# 연습문제1 : 리스트 '취미'에서 요소가 '커피', '차'이면 요소의 인덱스와 값을 출력한다.
# 선생님 답안3

# for 구문으로 리스트 favorites에 "커피"와 "차"가 포함되어 있는지 확인하고
# index 메서드로 그 요소의 인덱스를 구해 출력

for item in ('커피', '차'):
    if item in favorites:
        index = favorites.index(item)
        print(f'{index}. {item}')
```

```python
# 박준현씨 질문
# for, if 바꿔서 써도 상관이 없나요?

if item in favorites:
    for item in ('커피', '차'):
        index = favorites.index(item)
        print(f'{index}. {item}')
```

```python
# 연습문제2 : 리스트 취미에서 요소가 '콜라'라면 '소프트 드링크'로 출력하고 그렇지 않으면 요소를 그대로 출력한다.
# 내 답안

favorites = ['커피', '차', '쥬스', '콜라', '우유']

for i in range(len(favorites)):
    if favorites[i] == '콜라':
        print(f'{i}.', '소프트 드링크')
    else:
        print(f'{i}.', favorites[i])
```

```python
# 연습문제2 : 리스트 취미에서 요소가 '콜라'라면 '소프트 드링크'로 출력하고 그렇지 않으면 요소를 그대로 출력한다.
# 선생님 답안1

# for 구문과 range 함수로 리스트 favorites의 요소를 차례대로 확인
for i in range(len(favorites)):
    if favorites[i] == '콜라':
        print(f'{i}.', '소프트 드링크')
    else:
        print(f'{i}.', favorites[i])
```

```python
# 연습문제2 : 리스트 취미에서 요소가 '콜라'라면 '소프트 드링크'로 출력하고 그렇지 않으면 요소를 그대로 출력한다.
# 선생님 답안2

# for 구문과 enumerate 함수로 리스트 favorites의 요소를 차례대로 확인
for index, item in enumerate(favorites):
    # 인덱스가 변수 i인 요소가 "커피"이거나 "차"이면 출력
    if item == '콜라':
        print('소프트 드링크')
    else:
        print(item)
```

```python
# 연습문제2 : 리스트 취미에서 요소가 '콜라'라면 '소프트 드링크'로 출력하고 그렇지 않으면 요소를 그대로 출력한다.
# 선생님 답안3

# for 구문으로 리스트 favorites의 요소를 차례대로 확인
for item in favorites:
    if item == '콜라':
        print('소프트 드링크')
    else:
        print(item)
```

```python
# 연습문제2 : 리스트 취미에서 요소가 '콜라'라면 '소프트 드링크'로 출력하고 그렇지 않으면 요소를 그대로 출력한다.
# 선생님 답안4

# for 구문으로 리스트 favorites의 요소를 차례대로 확인
for item in favorites:
    # 조건부 표현식으로 변수 item이 '콜라'이면 '소프트 드링크'를 출력
    print('소프트 드링크' if item == '콜라' else item)
```

```python
# 연습문제2 : 리스트 취미에서 요소가 '콜라'라면 '소프트 드링크'로 출력하고 그렇지 않으면 요소를 그대로 출력한다.
# 선생님 답안5

# 빈 리스트 생성
copy = []

# for 구문으로 리스트 copy에 리스트 favorites를 복사
for item in favorites:
    copy.append('소프트 드링크' if item == '콜라' else item)
    
# 출력
print(copy) # 그냥 출력
print(*copy) # 언패킹 연산자 사용
print(*copy, sep=' / ') # 언패킹 연산자 사용 + 대행 연산자
print(*copy, end=' / ') # 언패킹 연산자 사용 + 대행 연산자
```
