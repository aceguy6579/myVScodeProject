```python
# 연습문제 : 사용자에게 자연수 n을 입력받은 뒤 자졍수 1의 자리 숫자, 10의자리 숫자, 100의 자리 숫자 등을 구해 출력한다.
# 내 답안

n1 = int(input('Enter two positive integers: '))

print(f'n1 = {n1}')

print('1의 자리 = ', n1 % 10)
print('10의 자리 = ', (n1 % 100 - (n1 % 10)) / 10)
print('100의 자리 = ', (n1 % 1000 - (n1 % 100)) / 100)
print('1000의 자리 = ', (n1 % 10000 - (n1 % 1000)) / 1000)
```
