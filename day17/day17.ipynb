```python
# 숙제 : 2026년 1월 1일은 목요일이다. 사용자에게 "월"과 "일"을 정수로 입력받아, 요일을 구해 출력

# 각 달의 일수를 튜플에 저장
MONTH_DAYS = (0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31)

# 사용자 입력
month = int(input('Enter the month as a number: '))
day = int(input('Enter the day as a number: '))

# 사용자가 입력한 날짜를 검사(validation)
if month < 1 or month > 12:
    print('ERROR: Invalid month.')
    raise SystemExit

if day < 1 or day > MONTH_DAYS[month]:
    print('ERROR: Invalid day.')
    raise SystemExit

# 2026년 1월 1일은 목요일
DAY_WEEK_FIRST = 4

# 사용자가 입력한 날짜가 1월 1일에서 몇일 지났는지 나타내는 변수
days = 0

# 사용자가 입력한 날짜가 1월 1일에서 몇일 지났는지 계산

# for 구문으로 1월에서부터 사용자가 입력한 "월"의 직전 월까지의 날짜를 변수 days에 누적해서 더한다.
for i in range(1, month):
    days += MONTH_DAYS[i]

# 사용자가 입력한 "일"을 변수 days에 더하고, 1을 뺀다.
days += day - 1

# 사용자가 입력한 날짜의 요일을 계산
day_week = (DAY_WEEK_FIRST + days) % 7

# 출력
print(f'2026-{month}-{day} {day_week}')
```
